```r
############## Exercise 5 ################
# HR department gives us a list of names for the participants
names = c('John', 'Suzie', 'Lestor', 'Sarah', 'Kitty', 'June', 'Kenneth', 'Aloysius', 'Angie', 'Zack', 'Stephen', 'Peter')
dept = c('HR', 'Sales', 'Sales', 'Sales', 'Sales', 'Sales', 'HR', 'HR', 'Sales', 'HR', 'Sales', 'HR')

# Given that we have so many arrays, and each of them are ordered 
#according to their positions (i.e.
# the first positions of each array gives the data for 1 entity: 
# ('John', 'male', 1.7, 'HR'))
# To make things easier to manipulate, we decide to make it into a table.
# In 1 line of R, construct a dataframe from the series
heights_df = data.frame(cbind(names, gender, dept, heights))

# Check the datatype of your dataframe. Is heights in numeric form?
str(heights_df)

# convert heights to numeric
# 1. use existing heights column, convert that to numeric 
# (this takes up a new memory address in
# the computer's ram, distinct from the existing heights column) and 
# 2. reassign it back to the heights column in the dataframe 
# (this replaces the existing heights 
# column in the memory with the data stored in the new memory address.)

# Idiosyncracy of R: 
# 1. Since heights is structured as factor in the dataframe, 
#    using as.numeric will not
#    convert the values of the levels, but the levels itself 
#    (i.e. re-indexed from 1, 2, 3, ...)
# 2. To solve this problem, we can call the values of the levels 
#    using the levels function
#    levels(dataframe$col) will give u the values of the levels, 
#    and we can convert this
#    array of strings into numeric and assign it back to the 
#    original column
levels_from_heights = levels(heights_df$heights)  # no repeat values
# this is a bug, we need to solve this.
heights_df$heights = as.numeric(levels_from_heights)[heights_df$heights]  # This is the correct way
# Check ?factor

# Out of curiosity, we would like to see if there is an association 
# between gender and department.
# Hint: create a table where columns are departments and rows are gender
# check out the function xtabs (?xtabs)
freq_table = xtabs(~ gender + dept, data=heights_df)

# construct a test of independence 
test = chisq.test(freq_table)

# HR says that they want to make uniform for the participants. 
# They only want the number of orders
# for Large, Medium and Small.
# The sizing chart goes like this:
# Height < 1.5m : Small
# Height between 1.5m and 1.7m : Medium
# Height > 1.7m: Large
# Count the number of uniform sizes
# Hint: sum(FALSE, TRUE, FALSE, TRUE) will return the number of TRUE, 
# in this case 2.

no_of_large = sum(heights_df$heights < 1.5)
no_of_medium = sum((1.5 <= heights_df$heights) & (heights_df$heights < 1.7))
no_of_small = sum(heights_df$heights > 1.7)

# HR would also like our help to compute the delivery fees of the uniform.
# The fees schedule is as follows:
# Just 1 piece: 2.40 each
# more than 1 piece but less than 3: 2.00 each
# more than 3 pieces: 1.50 each
# Use an if-elseif-else statement to calculate the 
# total delivery fees for medium-size uniforms.
# Hint: create a variable to capture the delivery fees and first 
#       assign it with 0.
#       As we compute the fees, we can add to this variable incrementally.
#       The code to do so is delivery_fees = delivery_fees + computed_fees
#       R processes the instruction by creating a new variable (internally) for
#       the right-hand-side (delivery_fees + computed_fees) and 
#       reassigns it to
#       the left-hand-side (delivery_fess). The value of delivery_fees thus increases
#       with each iteration.

delivery_fees = 0

# Perform an if-elseif-else statement for no_of_medium
if (no_of_medium == 1){
  delivery_fees = delivery_fees + 2.40 * no_of_medium
} else if (no_of_medium > 1 & no_of_medium < 3) {
  delivery_fees = delivery_fees + 2.0 * no_of_medium
} else {
  delivery_fees = delivery_fees + 1.50 * no_of_medium
}
```

```r
############## Exercise 6 ################
# We then realized that we should update the dataframe 
# who should receive what uniform sizes.
# This will make it easy for us to distribute them when the 
# uniforms arrive.
# Use a for loop to do this
# Hint #1: Use an unique identifier (i.e. Names) and iterate through it
# Hint #2: New columns can be created on-the-fly by assigning a value 
#          to a new column name
#          heights$uniform_size = 'Small", will create a new column 
#          called uniform_size
#          with the value of 'Small".
# Hint #3: For Hint #2, the entire column will have 1 value: 'Small". 
#          We can however, use
#          use a condition (heights_df$names == name) to extract 
#          just the corresponding
#          position of uniform_size (i.e. heights$uniform_size[condition] 
#          where our condition
#          is based on a unique identifier, will only return 1 row). 
#          Assigning a value to this will not affect other rows.

names = heights_df$names

for (name in names){
  h = heights_df$heights[heights_df$names == name]
  if (h < 1.5){
    heights_df$uniform_size[heights_df$names == name] = 'Small'
  } else if (h >- 1.5 & h < 1.7){
    heights_df$uniform_size[heights_df$names == name] = 'Medium'
  } else {
    heights_df$uniform_size[heights_df$names == name] = 'Large'
  }
}
```
