```python
import pandas as pd

# We will use the Wisconsin Breast Cancer dataset
# This dataset contains 569 samples of malignant and benign tumor cells
# First 2 columns are unique IDs with corresponding diagnosis (M=malignant and B=benign)
# Subsequent columns contain 30 real-value variables that were computed from digitized images of
# the cell nuclei

# Load data into pandas dataframe
df = pd.read_csv('wdbc.csv', header=None)
df
```

```python
# Separate dataset into outcome (y) and predictors (X)
from sklearn.preprocessing import LabelEncoder
X = df.loc[:, 2:].values
y = df.loc[:, 1].values

# Need to transform the class labels into integers
# To accomplish this easily, we use the LabelEncoder from scikit-learn
le = LabelEncoder()  # Call the LabelEncoder class
y = le.fit_transform(y)  # Fit and transform
```

```python
# Check which class is labeled as what
le.transform(['M', 'B'])  # Malignant is 1, Benign is 0
```

```python
# Divide the dataset into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
# As emphasized multiple times, most ML algorithms require the same scale for their input features
# for optimal performance
# Subsequent to that, we would like to reduce the dimensionality using PCA
# After which, we will run a Logistic regression on the data

# This is how we can use a pipeline to string this entire process up
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe_lr = Pipeline([
        ('scl', StandardScaler()),
        ('pca', PCA(n_components=2)),
        ('clf', LogisticRegression(random_state=1))
])  # Pipeline takes tuples of arguments. First element of each tuple is an identifier string.

pipe_lr.fit(X_train, y_train)
print('Test Accuracy: {:.3f}'.format(pipe_lr.score(X_test, y_test)))  # The accuracy is quite impressive.
# Using models within the open source scikit-learn and data, we can predict breast cancer with an accuracy of 94.7%
```

```python
# Using cross-validation to assess model performance (Slide 53 onwards)
import numpy as np
from sklearn.cross_validation import StratifiedKFold

kfold = StratifiedKFold(
    y=y_train,
    n_folds=10,
    random_state=1
)

scores = []

for k, (train, test) in enumerate(kfold):
    pipe_lr.fit(X_train[train], y_train[train])
    score = pipe_lr.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: {}, Class dist.: {}, Acc: {:.3f}'.format(k+1, np.bincount(y_train[train]), score))
```

```python
# Statistical distribution of accuracy
print('CV accuracy: {:.3f} +/- {:.3f}'.format(np.mean(scores), np.std(scores)*1.96))

# Our model's mean accuracy is 95%, with a standard deviation of 5.7%
```

```python
# Doing it in one pass with scikit-learn
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(
    estimator=pipe_lr,
    X=X_train,
    y=y_train,
    cv=10,
    n_jobs=4  # This is the parallelization parameter, specify number of cores to use
)

print('CV accuracy scores: {}'.format(scores))
print('CV accuracy: {:.3f} +/- {:.3f}'.format(np.mean(scores), np.std(scores)*1.96))
```

```python
# Performance diagnostics - learning and validation curves (Slide 56)
# Learning curves
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

%matplotlib inline
```

```python
# Refer to Slide 56 for the diagram
pipe_lr = Pipeline([
        ('scl', StandardScaler()),
        ('clf', LogisticRegression(
            penalty='l2', random_state=0))
    ])

train_sizes, train_scores, test_scores = learning_curve(
    estimator=pipe_lr,
    X=X_train,
    y=y_train,
    train_sizes=np.linspace(0.1, 1.0, 10),  # 10 evenly spaced relative intervals
    cv=10,
    n_jobs=4
)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.plot(train_sizes, train_mean,
         color='blue', marker='o',
         markersize=5,
         label='training accuracy')
plt.fill_between(train_sizes,
                 train_mean + train_std,
                 train_mean - train_std,
                 alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean,
         color='green', linestyle='--',
         marker='s', markersize=5,
         label='validation accuracy')
plt.fill_between(train_sizes,
                 test_mean + test_std,
                 test_mean - test_std,
                 alpha=0.15, color='green')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.8, 1.0])

# Potential signs of overfitting
```

```python
# Validation curves (Slide 56)
from sklearn.model_selection import validation_curve
param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
train_scores, test_scores = validation_curve(
               estimator=pipe_lr,
                X=X_train,
                y=y_train,
                param_name='clf__C',  # To access the parameter C in the LogisticRegression class
                param_range=param_range,
                cv=10)
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.plot(param_range, train_mean,
         color='blue', marker='o',
         markersize=5,
         label='training accuracy')
plt.fill_between(param_range, train_mean + train_std,
                 train_mean - train_std, alpha=0.15,
                 color='blue')
plt.plot(param_range, test_mean,
         color='green', linestyle='--',
         marker='s', markersize=5,
         label='validation accuracy')
plt.fill_between(param_range,
                 test_mean + test_std,
                 test_mean - test_std,
                 alpha=0.15, color='green')
plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])

# Sweet spot about 0.1
```

```python
# Searching for the best hyperparameters - Grid Search (Slide 57)
# The idea is to test the model on a range of hyperparameters and choose the best one

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
pipe_svc = Pipeline([('scl', StandardScaler()),
                     ('clf', SVC(random_state=1))])
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [
    {'clf__C': param_range,
     'clf__kernel': ['linear']},
    {'clf__C': param_range,
     'clf__gamma': param_range,  # RBF kernels have this additional parameter
     'clf__kernel': ['rbf']}]

gs = GridSearchCV(
    estimator=pipe_svc,
    param_grid=param_grid,
    scoring='accuracy',
    cv=10,
    n_jobs=-1)  # Uses all cores

gs = gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)

# Note that this is computationally very expensive, another alternative method is randomized search
# which draws random samples from the dataset
```

```python
# Use the best parameters on the test set
clf = gs.best_estimator_
clf.fit(X_train, y_train)

print('Test accuracy: {:.3f}'.format(clf.score(X_test, y_test)))  # Accuracy improves further
```

```python
# We can perform a hyperparameter tuning + performance diagnostic
# First perform a Grid Search CV to tune parameters
gs = GridSearchCV(
    estimator=pipe_svc,
    param_grid=param_grid,
    scoring='accuracy',
    cv=10,
    n_jobs=-1  # Uses all cores
)  

# Followed by a second cv the usual way
scores = cross_val_score(gs, X, y, scoring='accuracy', cv=5)

print('CV accuracy: {:.3f} +/- {:.3f}'.format(np.mean(scores), np.std(scores)))
```

```python
# Compare the above results to a simple decision tree
from sklearn.tree import DecisionTreeClassifier
gs = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=0),
    param_grid=[
        {'max_depth': [1, 2, 3, 4, 5, 6, 7, None]}
    ],
    scoring='accuracy',
    cv=5)

scores = cross_val_score(gs, X, y, scoring='accuracy', cv=5)

print('CV accuracy: {:.3f} +/- {:.3f}'.format(np.mean(scores), np.std(scores)))
```

```python
# How does it perform against a random forest?
from sklearn.ensemble import RandomForestClassifier
gs = GridSearchCV(
    estimator=RandomForestClassifier(
        criterion='entropy',
        random_state=1,
        n_jobs=4),
    param_grid=[
        {'n_estimators': [10, 50, 100]}
    ],
    scoring='accuracy',
    cv=5
)

scores = cross_val_score(gs, X, y, scoring='accuracy', cv=5)

print('CV accuracy: {:.3f} +/- {:.3f}'.format(np.mean(scores), np.std(scores)))
```

```python
# Confusion matrix (Slide 58)
from sklearn.metrics import confusion_matrix
pipe_svc.fit(X_train, y_train)
y_pred = pipe_svc.predict(X_test)
cmatrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(cmatrix)
```

```python
# Visualize the confusion matrix
fig, ax = plt.subplots(figsize=(2.5, 2.5))
ax.matshow(cmatrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(cmatrix.shape[0]):
    for j in range(cmatrix.shape[1]):
        ax.text(x=j, y=i,
                s=cmatrix[i, j],
                va='center', ha='center')
plt.xlabel('predicted label')
plt.ylabel('true label')
```

```python
# Other scoring metrics individually called
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score, f1_score
print('Precision: {:.3f}'.format(precision_score(y_true=y_test, y_pred=y_pred)))
print('Recall: {:.3f}'.format(recall_score(y_true=y_test, y_pred=y_pred)))
print('F1: {:.3f}'.format(f1_score(y_true=y_test, y_pred=y_pred)))

# Check out scikit-learn documentation for more metrics
```
